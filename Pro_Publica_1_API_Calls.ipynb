# API Calls to ProPublica to gather data

1. To avoid server loads and timeout issues the API calls were split by Congress and bill types
2. Multiple API calls were made for the below data sets and saved into multiple csv files. 
Congress 112 & bill type 'hr'
Congress 112 & bill type 's'
Congress 113 & bill type 'hr'
Congress 113 & bill type 's'
Congress 114 & bill type 'hr'
Congress 114 & bill type 's'
Congress 115 & bill type 'hr'
Congress 116 & bill type 'hr'
Congress 116 & bill type 's'
3. CSV file naming convention: "output_data/bills_{bill_type}_{congress}.csv"
Eg. "bills_s_112.csv"

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
import json
from json.decoder import JSONDecodeError

# Import API key
from api_keys import propublica_api_key

# Output File (CSV)
output_data_file = "output_data/bills_1.csv"
```

```python
record_num1=1
congress=["114"]
bill_slug = []
congress1 = []
for k in range(0,len(congress)):
    for i in range(1,10000,1):
        try:
            bill_id=f"s{i}"
            query_url = f"https://api.propublica.org/congress/v1/{congress[k]}/bills/{bill_id}.json"
            header = {"X-API-Key": propublica_api_key}
            response=requests.get(query_url , headers=header).json()
            bill_slug.append(response['results'][0]['bill_slug'])
            congress1.append(congress[k])
        except (KeyError, IndexError):
            print(f"{bill_id}: Bill not found. Skipping...")
        except JSONDecodeError as exc:
            #if exc.msg == 'Invalid \\escape':
                #text = text[:exc.pos] + '\\' + text[exc.pos:]
            
            print(f"{bill_id}: {exc.msg}")
        except:
            print ("Unexpected error")
       #If statement to handle the API call limits
        if record_num1 == 500:
            record_num1 += 1
            time.sleep(10)
            
```

```python
bills_df=pd.DataFrame({
                    "bill_slug":bill_slug,
                    "congress":congress1,
                    "bill_id":"" ,
                    "bill_type":"" ,   
                    "bill_uri":"" ,
                    "short_title":"" , 
                    "sponsor_title":"" ,
                    "sponsor_id":"" , 
                    "sponsor_name":"" ,
                    "sponsor_state":"" ,
                    "sponsor_party":"" ,
                    "congressdotgov_url":"" ,
                    "govtrack_url":"" ,
                    "introduced_date":"" ,
                    "active":"" ,
                    "house_passage":"" ,
                    "senate_passage":"" ,
                    "enacted":"" ,
                    "vetoed":"" ,
                    "cosponsors":"" ,
                    "cosponsors_by_party_D":"" ,
                    "cosponsors_by_party_R":"" ,
                    "committees":"" ,
                    "committee_codes":"" ,
                    "primary_subject":""   
                    })
bills_df
```
