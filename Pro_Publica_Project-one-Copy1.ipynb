```python
record_num1=1
congress=["115"]
bill_slug = []
congress1 = []
for k in range(0,len(congress)):
    for i in range(1,15000,1):
        try:
            bill_id=f"hr{i}"
            query_url = f"https://api.propublica.org/congress/v1/{congress[k]}/bills/{bill_id}.json"
            header = {"X-API-Key": propublica_api_key}
            response=requests.get(query_url , headers=header).json()
            bill_slug.append(response['results'][0]['bill_slug'])
            congress1.append(congress[k])
        except (KeyError, IndexError):
            print(f"{bill_id}: Bill not found. Skipping...")
        except JSONDecodeError as exc:
            #if exc.msg == 'Invalid \\escape':
                #text = text[:exc.pos] + '\\' + text[exc.pos:]
            
            print(f"{bill_id}: {exc.msg}")
        except:
            print ("Unexpected error")
       #If statement to handle the API call limits
        if record_num1 == 500:
            record_num1 += 1
            time.sleep(10)
            
```
